Working with large amounts of data

```python
new_df = pd.DataFrame(columns = df.columns)

for df in pd.read_csv('modified.csv', delimiter = "\t", chunksize=5):
    results = df.groupby(['Type 1']).count()

    new_df = pd.concat([new_df, results])

new_df
```
